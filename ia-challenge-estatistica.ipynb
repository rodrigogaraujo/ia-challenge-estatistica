```python
import pandas as pd
import matplotlib.pyplot as plt

# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}

df = pd.DataFrame(dict_faturamento)
```

```python
df.valor.mean()
```

```python
ax = df.plot.bar(x='data_ref', y='valor', legend=False, title='Faturamento por Data de Referência', color='skyblue', 
            figsize=(12, 6))
ax.bar_label(ax.containers[0])
```

```python
ax = df.plot.line(x='data_ref', y='valor', legend=False, title='Faturamento por Data de Referência', color='skyblue', 
            figsize=(12, 6))

ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'R$ {x:,.0f}'))
```
